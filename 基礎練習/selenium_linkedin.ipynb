```python
#　載入selenium相關套件
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import json
from pathlib import Path
base_path = Path(r'%notebook').absolute().parent.parent.parent
with open(fr"{base_path}\config.json", "r") as f:
    config = json.load(f)
```

```python
# 設定chrome driver的執行入徑
options = Options()
options.chrome_executable_path = config['path']['chromedriver']
```

```python
# 建立driver物件實體
driver = webdriver.Chrome(options = options)
# 連線到linkedin
driver.get("https://www.linkedin.com/jobs/search?trk=guest_homepage-basic_guest_nav_menu_jobs&position=1&pageNum=0")
# 捲動視窗，等待瀏覽起得出更多資料
n=0
while n<3:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)
    n+=1

# 抓出文章標題
title_tags = driver.find_elements(By.CLASS_NAME, "base-search-card__title") #搜尋class屬性是title的所有標籤
for tag in title_tags:
    print(tag.text)

driver.close()
```
